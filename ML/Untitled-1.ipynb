## Коррекция типов данных

```python
data['carat'] = (data['carat'] * 100).astype(int)
data['depth'] = (data['depth'] * 10).astype(int)
data['table'] = data['table'].astype(int)
data['x'] = (data['x']*100).astype(int)
data['y'] = (data['y']*100).astype(int)
data['z'] = (data['z']*100).astype(int)
data.info()
```

```python
data['cut'].value_counts()
```

```python
dict_cut = {
    'Ideal' : 4,
    'Premium' : 3,
    'Very Good' : 2,
    'Good' : 1,
    'Fair' : 0,
}
data['cut'] = data['cut'].map(dict_cut)
```

```python
def get_border_data(dataframe, colomn):
    Q1= dataframe[colomn].quantile(0.25)
    Q3 = dataframe[colomn].quantile(0.75) 

    IQR = Q3 - Q1 

    low_bound = Q1 - 1.5 * IQR
    up_bound = Q3 + 1.5 * IQR
    
    return [low_bound, up_bound]
```

```python
def filtered(dataframe, *args):
    filtered_df = dataframe.copy()
    for column in args:
        low_bound, up_bound = get_border_data(filtered_df, column)
                
        filtered_df = filtered_df[
            (filtered_df[column] >= low_bound) & 
            (filtered_df[column] <= up_bound)]
    
    return filtered_df
```

```python
filteredata = filtered(data, 
                       'x',
                       'y',
                       'z',
                       'table',
                       'carat',
                       'depth',
                       'price')
```
