```python
def analyzing(data, colomns):
    data_melted = data.melt(value_vars=colomns, var_name='Colomn', value_name='Value')

    plt.figure(figsize=(12, 9))
    sns.boxplot(x='Colomn', y='Value', data=data_melted)
    plt.grid(True)
    plt.show()
analyzing(data, colomns =['x', 'y', 'z'])
```

```python
analyzing(data, 'price')
```

```python
analyzing(data, colomns=['carat', 'table'])
```

```python
analyzing(data, 'depth')
```

На основе boxplot, построенных для характеристик бриллиантов. Видно, что большинство значений сосредоточено в определённом диапазоне. Однако, есть выбросы, которые могут быть связаны с уникальными, аномальными или ошибочными экземплярами. Подкорректируем значения

```python
def get_border_data(dataframe, colomn):
    Q1= dataframe[colomn].quantile(0.25)
    Q3 = dataframe[colomn].quantile(0.75) 

    IQR = Q3 - Q1 

    low_bound = Q1 - 1.5 * IQR
    up_bound = Q3 + 1.5 * IQR
    
    return [low_bound, up_bound]
```

```python
def filtered(dataframe, args):
    filtered_df = dataframe.copy()
    for column in args:
        low_bound, up_bound = get_border_data(filtered_df, column)
                
        filtered_df = filtered_df[
            (filtered_df[column] >= low_bound) & 
            (filtered_df[column] <= up_bound)]
    
    return filtered_df
```

Реализована функция для фильтрации выбросов в данных на основе межквартильного размаха (IQR). Это позволяет удалить аномальные значения, которые могут искажать результаты анализа.

```python
filteredata = filtered(data, args= ['x','y','z','table','carat','depth','price'])
```

```python
filteredata.info()
```

```python
filteredata.hist(bins=50, figsize=(20,15))
plt.show()
```

Эти гистограммы позволяют визуализировать распределение различных характеристик бриллиантов. Чтобы мы могли понимать в каком диапазоне работаем

```python
filteredata['v'] = filteredata['x'] * filteredata['y'] * filteredata['z']
```

```python
sns.histplot(filteredata['v'], bins=30, kde=True)
plt.title('Распределение объема бриллианта')
plt.show()
```

Был добавлен новый столбец v, который представляет собой объем бриллианта, рассчитанный как произведение длин по осям x, y и z. Это позволяет удобнее оценить кореляцию размерности каждого бриллианта в датасете.

Гистограмма с ядерной оценкой плотности (KDE) показывает, что большинство объемов сосредоточено в диапазоне от 0.5 до 2.5, что указывает на то, что большинство бриллиантов имеют относительно небольшие объемы.

```python
print (data.hist('price', bins=100))
print (filteredata.hist('price', bins=100))
```

На первой гистограмме видно, что распределение цен до фильтрации имеет длинный хвост в сторону более высоких цен. Бриллиантов с очень высокой ценой могут искажать общую картину.

На второй гистограмме, после фильтрации выбросов, распределение цен становится более компактным. Большинство цен сосредоточено в диапазоне до 10,000 единиц, что делает данные более репрезентативными для анализа.

```python
sns.lmplot(x='v', y='price', data=filteredata)
```

График lmplot показывает линейную зависимость между объемом бриллианта (v) и его ценой (price). Это позволяет визуализировать, как увеличение объема влияет на стоимость бриллианта.

Виден положительный тренд, указывающий на то, что с увеличением объема бриллианта его цена также увеличивается. Это ожидаемо, так как более крупные бриллианты обычно стоят дороже.

Несмотря на общий тренд, наблюдается значительный разброс данных. Это может быть связано с влиянием других факторов, таких как качество огранки, цвет, чистота и вес, которые могут также влиять на цену.

```python
sns.lmplot(x='v', y='carat', data=filteredata)
```

Определив линейную зависимость между объемом бриллианта и его весом, поможет нам с пониманием в анализе, так как вес является одним из ключевых факторов, влияющих на цену.
